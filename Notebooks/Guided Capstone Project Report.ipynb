# Guided Capstone Project Report

For the guided capstone project, I took on the role of a data scientist working for a ski resort called Big Mountain Resort, located in Montana.  The resort offers incredible views of Glacier National Park and Flathead National Forest, along with access to 105 trails.  Roughly 350,000 people of all skill levels visit the resort every year to ski and snowboard.  The mountain is serviced by 11 lifts, 2 t-bars, and a magic carpet.  Highest elevation is 6,817 ft and a base of 4,464 ft, vertical drop of 2,353 feet.  Recently Big Mountain installed an additional chair lift to help increase the distribution of visitors across the mountain.  This additional chair increases Big Mountains’s operating costs by $1.54m.  Currently the resort’s strategy is to charge a premium above the average price of resorts in its market segment, but there is suspicion that Big Mountain isn’t capitalizing on its own facilities as much as it could.  The business has hired me to implement a more data driven business strategy, with guidance on how to select a better value for their ticket price.

I created a problem statement worksheet to get the ball rolling on what should be expected for this problem.  How can Big Mountain select a better ticket price before the start of next ski season, by utilizing data from 330 other resorts, and by comparing its own facilities to other resorts.  In addition, Big Mountain will look to increase efficiencies and cut costs across the resort, and generate at least an additional $1.54M to cover the increased operating costs for the new ski lift.  Some constraints to be aware of are potentially failure to decrease costs and increase efficiencies, or potential pushback from stakeholders on the suggested new price for passes.  I am working with the director of operations, Jimmy Blackburn, as well as Alesha Eisen, the database manager.

Following this, I took some proposed scenarios from corporate and modeled them to show what my model would produce with those proposals.  Initially the mountain looked at closing the 10 least popular runs, and you can see that each run closure was more and more unfavorable to total revenue. 

The second scenario proposed by corporate looked to increase runs by 1, increase chairs by 1, and increase the vertical drop by 150 ft.  This was favorable to the tune of an increased ticket cost of 92 cents.  Over the season this would expect to gross an additional 1.6m.  Even though this seems substantial, it only barely covers the operating cost of the new lift.  Big mountain needs to make at least an extra 1.54m just to break even.  The third scenario added 2 acres of snow making are on top of scenario 3, and ended up with the same gross revenue gain of 1.6M  The 4th scenario looked to make the longest run .2 miles longer and increased the snow making acreage by 4 - this did not result in an increase in revenue.  

I found that the best scenario proposed by corporate was the second one.  Even though this only suggested a 92 cent increase, it still covers the additional costs of operating the new lift.  My respose back to corporate is to increase the ticket price by at least 5 dollars a ticket.  This is still only an increase of 50 percent of the suggested price hike that the model suggested (10.26 dollars).  Raising ticket prices by 5 dollars would bring the total ticket price to 86 dollars and would suggest a roughly 8.75m dollar gross revenue increase.  This is calculated by the following:  5 average days of skiing per visitor * ~350,000 visitors * ticket price increase of 5 dollars.  There is some room for ticket prices to increase even futher, and a full 10 dollars above (still supported by the model) would suggest an additional 17.5m dollars of revenue.  While this decision is ultimately up to the board and the execs to figure out, the data I have brought forth gives them reason to increase prices by a relatively small amount and increase revenue tremendously.  This could provide opportunities to update/expand/improve current facilities into even better ones, but that's far into the future and requires some additional analysis. I hope the analysis provides them with valuable information that can be utilized for the upcoming seasons.
